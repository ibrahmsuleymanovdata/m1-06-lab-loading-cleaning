# Task 2

```python
tickets['opened_at'] = pd.to_datetime(tickets['opened_at'])
```

```python
tickets.info()
```

```python
tickets.isnull().sum()
```

```python
tickets['category']=tickets['category'].str.strip().str.lower()
```

```python
tickets['priority']=tickets['priority'].str.strip().str.lower()
```

```python
print(tickets['category'].unique())
print(tickets['priority'].unique())
```

# Task 3

```python
tickets["resolution_minutes"] = pd.to_numeric(tickets["resolution_minutes"], errors="coerce")
```

```python
tickets.info()
```

```python
tickets.isnull().sum()
```

```python
assert tickets["resolution_minutes"].isnull().sum() == 100
print("True!")
```

# Task 4

```python
tickets_clean = tickets.dropna(subset=['resolution_minutes'])
```

```python
tickets_clean
```

```python
(tickets["resolution_minutes"] - tickets_clean["resolution_minutes"]).unique()
```

```python
print("Unique Categories:", tickets_clean["category"].unique())
print("Unique Priorities:", tickets_clean["priority"].unique())
```

```python
print(f"Records before cleaning: {len(tickets)}")
print(f"Records after cleaning: {len(tickets_clean)}")
print(f"Total rows removed: {len(tickets) - len(tickets_clean)}")
```

# Task 5


```python
dict_ = {"total_records":len(tickets),
         "clean_records": len(tickets_clean),
         "missing_resolution":tickets["resolution_minutes"].isna().sum(),
         "avg_resolution": tickets_clean["resolution_minutes"].mean(),
         "max_resolution": tickets_clean["resolution_minutes"].max(),                        
}
```

```python
dict_
```

```python
print("Quality Summary:")
for key, value in dict_.items():
    print(f"{key}: {value}")

is_verified = (dict_["clean_records"] + dict_["missing_resolution"]) == dict_["total_records"]

print(f"\nVerification (Clean + Missing == Total): {is_verified}")
```
